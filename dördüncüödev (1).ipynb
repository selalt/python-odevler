```python
#  Verilen liste
liste = ["Python", True, 9, "3", 8.4, "Hi-Kod", "False", 4.7]

#  3 ögesine ulaşmak için
print(liste[3])  

#  Hi-Kod, değerine ulaşmak için 
print(liste[5])  

#  4.7 değerine ulaşmak için
print(liste[7]) 

#  9,"3",8.4,"Hi-Kod" değerlerine ulaşmak için slicing
print(liste[2:6]) 

#  8.4,"Hi-Kod","False",4.7 değerlerine ulaşmak için slicing
print(liste[4:8])  

#  String veri tipindeki öğeleri yeni_liste'ye ekleme
yeni_liste = []
for item in liste:
    if isinstance(item, str):
        yeni_liste.append(item)
print("Stringler listesi:", yeni_liste)  


#  Enumerate örneği
meyveler = ["Elma", "Armut", "Muz", "Çilek"]

for index, meyve in enumerate(meyveler):
    print("{}. indexte bulunan meyve: {}".format(index, meyve))
```
