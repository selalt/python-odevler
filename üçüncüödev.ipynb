```python
#Soru1

def daire_alani(pi,r):
    alan = pi*(r**2)
    print("Dairenin alanı = {}".format(alan) )
    return alan

daire_alani(3.14, 5)
```

```python
#Soru2

def faktoriyel(n):
    sonuc = 1
    for i in range(1,n+1):
        sonuc *=i
    print("{}.sayısının faktoriyeli = {}".format(n,sonuc))
    return sonuc

faktoriyel(5)
```

```python
#Soru3

def yas_hesapla(dogum_yili):
    from datetime import datetime
    bu_yil = datetime.now().year
    yas= bu_yil-dogum_yili
    print(yas)
    return yas

yas_hesapla(1992)
```

```python
#Soru4

def emeklilik_hesapla(isim, dogum_yili):
    yas = yas_hesapla(dogum_yili)

    if yas >= 65:
        print("Emekli oldunuz.")
    else:
        kalan = 65 - yas
        print("{}, emekliliğine {} yıl kaldı.".format(isim, kalan))

emeklilik_hesapla("Selin",1992)
```
